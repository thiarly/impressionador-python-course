# Formas de interromper um for

### 2 Opções:

- break -> interrompe e finaliza o for
- continue -> interrompe e vai para o próximo item do for

```python
vendas = [100, 150, 1500, 2000, 120]
```

- Caso 1: Se todas as vendas forem acima da meta, a loja ganha bônus

```python
meta = 110

for venda in vendas:
    if venda < meta:
        print('Loja não ganha bônus. Vendeu 1 valor abaixo R$ {} da meta'.format(venda))
        break
    print(venda)
```

- Caso 2: Exiba quem bateu a meta

```python
vendedores = ['João', 'Julia', 'Ana', 'José', 'Maria']
meta = 130

for i, vendedor in enumerate(vendas):
    if vendedor >= meta:
        print(vendedores[i], vendas[i])
```

```python
#lógica, se a venda for menor q a meta, o código retorna do CONTINUE para o INICIO, se a venda for maior ele pula
# para o print.

for venda in vendas:
    if venda < meta:
        continue
    print(venda)
```
